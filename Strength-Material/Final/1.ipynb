```python
l = 15
from sympy import *
from sympy.physics.continuum_mechanics.beam import Beam
E, I = symbols('E, I')
R1, R2 = symbols('R1, R2')
bm = Beam(l, E, I)
bm.apply_load(R1, 0, -1)
bm.apply_load(20,0,0,end=6)
bm.apply_load(40,9,-1)
bm.apply_load(R2,12,-1)
bm.apply_load(20,l,-1)
bm.bc_deflection = [(0, 0), (12, 0)]
bm.solve_for_reaction_loads(R1, R2)
bm.reaction_loads
```

```python
bm.max_bmoment()
```

```python
bm.plot_loading_results(subs={E:10e9, I:1.5e-6})
```
